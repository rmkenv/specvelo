# SpecVel — Quickstart

Run any adapter in ~5 lines. No API key needed for equities and commodities.

**Free FRED API key** (for fixed income + macro): https://fred.stlouisfed.org/docs/api/api_key.html

```python
import sys, os
sys.path.insert(0, os.path.join(os.getcwd(), '..', 'specvel'))

# Optional: set your FRED key here
FRED_KEY = os.environ.get('FRED_API_KEY', 'your_key_here')
```

## Equities (no key needed)

```python
from adapters.equities import EquitiesAdapter
from leaderboard import run_leaderboard, print_leaderboard

adapter = EquitiesAdapter()
df_eq = run_leaderboard(adapter, start='2023-01-01', end='2026-03-10',
                        asset_class='equities', top_n=15)
print_leaderboard(df_eq, title='EQUITIES VELOCITY LEADERBOARD')
```

```python
# Save results
import os
os.makedirs('../results', exist_ok=True)
from leaderboard import save_leaderboard
save_leaderboard(df_eq, '../results/equities_leaderboard.csv')
df_eq[['label','signal','conviction','vel_zscore','momentum','anomaly_flag']].head(10)
```
